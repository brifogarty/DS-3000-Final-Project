```python
unique_spoturl_arr = chart_crawl_df['Spotify_URL'].unique()
unique_title_arr = chart_crawl_df['Title'].unique()
```

```python
def get_tracks_multiple_urls(chart_crawl_df_arg):

    dup_urls_list = list()
    for title in unique_title_arr:
        title_df = chart_crawl_df_arg[['Title', 'Spotify_URL']][chart_crawl_df_arg['Title']==title]
        url_srs = title_df['Spotify_URL'].unique()
        num_urls = len(url_srs)

        if (num_urls > 1):
            dup_urls_list.append([title, num_urls, list(url_srs)])
        else:
            continue
            
    result_df_cols = ['Title', 'Num Spotify_URLs', 'Spotify_URLs Arr']
    return pd.DataFrame(dup_urls_list, columns=result_df_cols)
```

```python
mult_urls_df = get_tracks_multiple_urls(chart_crawl_df)
```

```python
mult_urls_df.head()
```

```python
def get_titles_mult_artists(chart_crawl_df_arg):

    dup_artists_list = list()
    for title in unique_title_arr:
        title_df = chart_crawl_df_arg[['Title', 'Artist']][chart_crawl_df_arg['Title']==title]
        artists_srs = title_df['Artist'].unique()
        num_artists = len(artists_srs)
        
        if (num_artists > 1):
            dup_artists_list.append([title, num_artists, list(artists_srs)])
        else:
            continue
            
    result_df_cols = ['Title', 'Num Artists', 'Artists Arr']
    return pd.DataFrame(dup_artists_list, columns=result_df_cols)
```

```python
mult_artists_df = get_titles_mult_artists(chart_crawl_df)
```
