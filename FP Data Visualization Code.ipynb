```python
import pandas as pd
import numpy as np
```

```python
audioFeatures = pd.read_csv("ml_dataset.csv")
```

```python
audioFeatures.drop(["Unnamed: 0"], axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# bar charts showing count of Top 200 songs within various ranges of features
```

```python
danceability = px.histogram(audioFeatures, x='danceability')
danceability.update_layout(title_text = 'Danceability in Spotify Top 200 Songs')
danceability.show()
```

```python
energy = px.histogram(audioFeatures, x='energy')
energy.update_layout(title_text = 'Energy in Spotify Top 200 Songs')
energy.show()
```

```python
key = px.histogram(audioFeatures, x='key')
key.update_layout(title_text = 'Key in Spotify Top 200 Songs')
key.show()
```

```python
loudness = px.histogram(audioFeatures, x='loudness')
loudness.update_layout(title_text = 'Loudness in Spotify Top 200 Songs',
                      xaxis_title_text = 'loudness (decibels)')
loudness.show()
```

```python
mode = px.histogram(audioFeatures, x='mode')
mode.update_layout(title_text = 'Mode in Spotify Top 200 Songs')
mode.show()
```

```python
speechiness = px.histogram(audioFeatures, x='speechiness')
speechiness.update_layout(title_text = 'Speechiness in Spotify Top 200 Songs')
speechiness.show()
```

```python
acousticness = px.histogram(audioFeatures, x='acousticness')
acousticness.update_layout(title_text = 'Acousticness in Spotify Top 200 Songs')
acousticness.show()
```

```python
instrumentalness = px.histogram(audioFeatures, x='instrumentalness')
instrumentalness.update_layout(title_text = 'Instrumentalness in Spotify Top 200 Songs')
instrumentalness.show()
```

```python
liveness = px.histogram(audioFeatures, x='liveness')
liveness.update_layout(title_text = 'Liveness in Spotify Top 200 Songs')
liveness.show()
```

```python
valence = px.histogram(audioFeatures, x='valence')
valence.update_layout(title_text = 'Valence in Spotify Top 200 Songs')
valence.show()
```

```python
tempo = px.histogram(audioFeatures, x='tempo')
tempo.update_layout(title_text = 'Tempo in Spotify Top 200 Songs',
                   xaxis_title_text = "tempo (beats per minute)")
tempo.show()
```

```python
duration_ms = px.histogram(audioFeatures, x='duration_ms')
duration_ms.update_layout(title_text = 'Duration in Spotify Top 200 Songs',
                         xaxis_title_text = 'duration (milliseconds)')
duration_ms.show()
```

```python
time_signature = px.histogram(audioFeatures, x='time_signature')
time_signature.update_layout(title_text = 'Time Signature in Spotify Top 200 Songs',
                            xaxis_title_text = "time signature (number of beats in a bar)")
time_signature.show()
```

```python
fig = px.scatter_matrix(audioFeatures,
    dimensions=["danceability", "speechiness", "acousticness", "liveness", "valence", "energy"], color="max_chart_rank")
fig.update_traces(diagonal_visible=False)
fig.show()
```
